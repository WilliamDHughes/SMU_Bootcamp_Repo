```python
school_name = school_data.school_name
school_type = school_data.type
total_students = len(student_data)
total_school_budget = school_data.budget.sum()
per_student_budget = (total_school_budget / total_students)
average_math = student_data.math_score.mean()
average_reading = student_data.reading_score.mean()
```

```python
df2 = school_data_complete.copy(deep=True)
df2.head()
```

```python
students = df2.groupby(['school_name', 'type']).student_id.count()
students.head()
```

```python
tot_budget = school_data.groupby(['school_name', 'type']).budget.sum()
tot_budget.head()
```

```python
per_student = tot_budget / students
per_student.head()
```

```python
math_average = round((df2.groupby(['school_name', 'type']).math_score.mean()), 2)
math_average                            
```

```python
reading_average = round((df2.groupby(['school_name', 'type']).reading_score.mean()), 2)
reading_average.head()
```

```python
passing_math = df2.loc[df2.math_score >= 70].groupby(['school_name', 'type']).student_id.count()
passing_math.head()
```

```python
passed_math = round(((passing_math / students) * 100), 2)
passed_math.head()
```

```python
passing_reading = df2.loc[df2.reading_score >= 70].groupby(['school_name', 'type']).student_id.count() * 100
passed_reading = round((passing_reading / students), 2)
passed_reading.head()
```

```python
passing_both = (df2.loc[(df2.reading_score >= 70) & (df2.math_score >= 70)].groupby(["school_name", "type"]).student_id.count() * 100)
passed_both = round((passing_both / students), 2)
passed_both.head()
```

```python
#build out summary table
```

```python
summ2 = pd.DataFrame()
summ2["Total Students"] = students
summ2["Total Budget"] = tot_budget
summ2["Per Student Budget"] = per_student
summ2["Average Math"] = math_average
summ2["Average Reading"] = reading_average
summ2["% Passing Math"] = passed_math
summ2["% Passing Reading"] = passed_reading
summ2["% Passing Both"] = passed_both
summ2.head()
```

```python
df2['passing_reading'] = df2['reading_score'] >= 70
df2['passing_math'] = df2['math_score'] >= 70
df2['passing_both'] = (df2['math_score'] >= 70) & (df2['reading_score'] >= 70)
df2
```

```python
df3 = df2.groupby("school_name").agg({"type": "first", "student_id": "count", "budget": "first", "math_score": "mean", "reading_score": "mean", "passing_math": "sum", "passing_reading":"sum", "passing_both": "sum"})
df3.head()
```

```python
df3["per_student"] = df3.budget / df3.student_id 
df3["perc_pass_math"] = round((df3.passing_math / df3.student_id), 2) * 100 
df3["perc_pass_reading"] = round((df3.passing_reading / df3.student_id), 2) * 100 
df3["perc_pass_both"] = round((df3.passing_both / df3.student_id), 2) * 100
df3.head()
```

```python
df4 = df3.drop(["per_student", "perc_pass_math", "perc_pass_reading"], axis=1)
df4.head()
```
